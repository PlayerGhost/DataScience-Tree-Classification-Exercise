```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
```

```python
dados = pd.read_csv("Tree-Classification.csv")

dados = dados.sample(frac=1)

dados
```

```python
y = dados["Target"]

X = dados.drop(columns=["Target"])

print(X)
y
```

```python
def getModel():
 return DecisionTreeClassifier(max_depth=6)
```

```python
X_train,X_teste,y_train,y_teste = train_test_split(X.values,y.values,test_size = 0.2)


model = getModel().fit(X_train,y_train)

model.score(X_teste,y_teste)
```

```python
cross_val = cross_val_score(getModel(), X, y, cv=10)

cross_val.mean()
```

```python


plt.figure(figsize=(250,120))
plot_tree(model,fontsize=50)
```

```python


predicoes = cross_val_predict(getModel(), X.values, y, cv=10)

cm = confusion_matrix(y,predicoes)

cm_df = pd.DataFrame(cm)

cm_df
```
